```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
sns.displot(df['Age'],kde=True)
```

```python
for i in df.columns:
  sns.displot(df[i],kde=True)
```

```python
df.groupby('Age').size().sort_values().plot(kind='bar',figsize=(10,20))
```

```python
df.groupby('HeartDisease').size().plot(kind='pie',autopct='%0.1f')
```

```python
df.groupby('Cholesterol').size().nlargest(10).plot(kind='pie',autopct='%0.1f')
```

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct='%0.1f')
```

```python
sns.violinplot(x=df['Age'],data=df)
```

```python
sns.violinplot(x=df['Age'], y=df['Sex'])
```

```python
sns.violinplot(x=df['Cholesterol'], y=df['Sex'])
```

```python
sns.violinplot(x=df['HeartDisease'], y=df['Sex'])
#0- no disease, 1-disease
```

```python
df1=df[['Age','RestingBP','Cholesterol','FastingBS','MaxHR']]
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

```python
#Jointplot with respect to ages
for i in df.columns:
  if i !='Sex' and i !='RestingECG' and i !='ExerciseAngina' and i !='ChestPainType' and i !='ST_Slope':
    sns.jointplot(x='Age',y=i,data=df,kind='hex')
```
